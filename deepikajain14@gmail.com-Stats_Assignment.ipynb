```python
# Question 3
poll=pd.read_csv("C:\\Users\\Deepika\\Desktop\\DSP\\DS\\Fundamental Stats & Stats tests\\Basic Statistics Assignment\\pollination.csv")

poll.head()
```

```python
# Question 4

#  H0: mean1 = mean2 = mean3 = mean4 = mean5 = mean6
#  H1: at least one is not equal

dvd=pd.read_csv("C:\\Users\\Deepika\\Desktop\\DSP\\DS\\Fundamental Stats & Stats tests\\Basic Statistics Assignment\\dvdplayer.csv")

dvd.head()
```

```python
Above_65 =  dvd[dvd['agegroup']== '65 and over']['dvdscore']
Btwn55_64 = dvd[dvd['agegroup']== '55-64']['dvdscore']
Btwn35_44 = dvd[dvd['agegroup']== '35-44']['dvdscore']
Btwn25_34 = dvd[dvd['agegroup']== '25-34']['dvdscore']
Btwn45_54 = dvd[dvd['agegroup']== '45-54']['dvdscore']
Under_25 =  dvd[dvd['agegroup']== 'Under 25']['dvdscore']
```

```python
stats.f_oneway(Above_65,Btwn55_64,Btwn35_44,Btwn25_34,Btwn45_54,Under_25)

# Null Hypothesis rejected as pvalue is less than 5%
```

```python
# Question 5
sample=pd.read_csv("C:\\Users\\Deepika\\Desktop\\DSP\\DS\\Fundamental Stats & Stats tests\\Basic Statistics Assignment\\sample_survey.csv")

sample.head()
```

```python
# 1)  HO: the vaiables are not related
#     H1: the variables are related 
    
t = pd.crosstab(sample.marital, sample.wrkstat, margins = True)
t
```

```python
stats.chi2_contingency(observed = t)

# as pvalue is <5, there null is false and marital status and labour force status are related
```

```python
# 2)  HO: the vaiables are not related
#     H1: the variables are related 
    
u = pd.crosstab(sample.marital, sample.degree, margins = True)
u
```

```python
stats.chi2_contingency(observed = u)

# as pvalue is <5, there null hypothesis is false and marital status and education status are related
```

```python
# 3) HO: the vaiables are not related
#    H1: the variables are related 
    
v = pd.crosstab(sample.marital, sample.rincome, margins = True)
v
```

```python
stats.chi2_contingency(observed = v)

# as pvalue is <5, there null hypothesis is false and marital status and education status are related
```
